```python
from google import genai
from google.genai import types
import os
from dotenv import load_dotenv
from IPython.display import display, Markdown
```

```python
load_dotenv()
```

```python
config = types.GenerateContentConfig(
    temperature=0.7,
    top_k=40,      
    top_p=0.95,    
)

client = genai.Client(api_key=os.getenv("API_KEY"))
```

```python
content = """
Ngày tháng: 15/03/2026
Tiêu đề: (Trống)
Nội dung: Hôm nay có một chuyện không vui xảy ra làm mình rất tức giận và mất kiểm soát lời nói với người khác. Tối về, mình đã phải ngồi tĩnh lặng lại, mở một bài giảng Phật pháp lên nghe và viết những dòng này vào nhật ký. Mình nhận ra cơn giận này bùng phát có lẽ một phần do cái tôi của mình lúc đó quá lớn, không chịu lắng nghe.
"""

prompt = """
Từ đoạn nhật ký sau, hãy trích xuất thông tin và trả về JSON theo schema:

Đoạn nhật ký: {content}

YÊU CẦU:
1. Đọc kỹ nội dung nhật ký
2. Trích xuất thông tin thực tế từ nội dung
3. Điền vào JSON template sau (KHÔNG ĐƯỢC ĐỂ TRỐNG):
{{
  "ngay_thang": (Ngày tháng từ nhật ký với format là DD/MM/YYYY)
  "tieu_de": (Nếu trống thì null, không thì là tiêu đề của nhật ký dạng text)
  "cam_xuc_chinh": (liệt kê các cảm xúc tìm được trong đoạn nhật ký, format là list, ngăn cách bằng dấu phẩy, tối đa 3 cảm xúc chính)
  "nguyen_nhan_cot_loi": (phân tích nguyên nhân từ nội dung, tối đa 30 - 40 từ)
  "goi_y_chanh_niem": (lời khuyên dựa trên tình huống, tối đa 30 - 40 từ)
}}

QUY TẮC VÀNG: 
- Tất cả các field phải được điền dữ liệu THỰC TẾ từ nhật ký
- Không được trả về template rỗng
- Chỉ trả về JSON, không thêm text khác
"""
```

```python
response = client.models.generate_content(
    model="models/gemini-2.5-flash",
    config=config,
    contents=[prompt.format(content=content)]
)
```

```python
display(Markdown(response.text))
```
