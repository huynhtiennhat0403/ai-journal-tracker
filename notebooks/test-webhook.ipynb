```python
import requests
```

```python
url = "http://localhost:1234/webhook/diary"

# Dữ liệu mẫu bóc 
data = {
    "ngay_thang": "15/03/2026",
    "tieu_de": "",
    "noi_dung": "Hôm nay có một chuyện không vui xảy ra làm mình rất tức giận và mất kiểm soát lời nói với người khác. Tối về, mình đã phải ngồi tĩnh lặng lại, mở một bài giảng Phật pháp lên nghe và viết những dòng này vào nhật ký. Mình nhận ra cơn giận này bùng phát có lẽ một phần do cái tôi của mình lúc đó quá lớn, không chịu lắng nghe."
}

# Gửi HTTP POST request đến Webhook
response = requests.post(url, json=data)

print("Status Code:", response.status_code)
print("Response:", response.text)
```
